```python
import seaborn as sns

FAMILY_MAPPING = {
    # OpenAI
    "gpt-3.5-turbo": "OpenAI",
    "gpt-3.5-turbo-0125": "OpenAI",
    # "gpt-4": "OpenAI",
    # "gpt-4-turbo-preview": "OpenAI",
    "gpt-4-turbo-2024-04-09": "OpenAI",
    "gpt-4o-2024-05-13": "OpenAI",
    "gpt-4-0125": "OpenAI",
    "gpt-4-0613": "OpenAI",
    # Google
    # "gemini-pro": "Google",
    "gemini-1.0-pro": "Google",
    "gemini-1.5-pro": "Google",
    # Anthropic
    "claude-3-sonnet": "Anthropic",
    "claude-3-opus": "Anthropic",
    "claude-3-haiku": "Anthropic",
    # Meta
    "llama-3-70b-chat": "Meta",
    "llama-3-8b-chat": "Meta",
    # Mistral
    "mistral-large": "Mistral",
    "mistral-medium": "Mistral",
    "mistral-small": "Mistral",
    "mixtral-8x7b": "Mistral",
    # Alibaba
    "qwen1.5-32B-Chat": "Alibaba",
    "qwen1.5-72B-Chat": "Alibaba",
    "qwen1.5-110B-Chat": "Alibaba",
    # DBRX
    "dbrx-instruct": "Databricks",
    # Cohere
    "command-r-plus": "Cohere",
    "command-r": "Cohere",
    # Special
    "council": "Council",
    "council (by majority vote)": "Council",
    "council (by majority vote)": "Council",
    "council (by mean pooling)": "Council",
    "council (no aggregation)": "Council",
    "council (no aggregation)": "Council",
    "human_generic": "Human Generic",
}

FAMILY_COLORS = {
    "OpenAI": "mediumseagreen",
    "Anthropic": "burlywood",
    "Mistral": "darkorange",
    "Google": "skyblue",
    "Meta": "magenta",
    "Cohere": "coral",
    "Cohere": "darkslategray",
    "Databricks": "orangered",
    "Alibaba": "slateblue",
    "Council": "gold",
    "Human Generic": "black",
}

def plot_arena_hard_elo_stats(stats, title, ax=None):
    # Mapping and error calculations
    stats["family"] = stats["model"].map(FAMILY_MAPPING)
    stats["lower_error"] = stats["score"] - stats["lower"]
    stats["upper_error"] = stats["upper"] - stats["score"]
    errors = stats[["lower_error", "upper_error"]].T.values

    # Create the plot if ax is not provided
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 8))

    # Increase font size globally
    sns.set_context("talk")

    # Create the bar plot
    sns.barplot(
        data=stats,
        y="model",
        x="score",
        hue="family",
        palette=FAMILY_COLORS,
        ax=ax
    )

    # Add error bars and text annotations
    for i, (score, error) in enumerate(zip(stats["score"], errors.T)):
        try:
            ax.errorbar(
                score, i, xerr=[[error[0]], [error[1]]], fmt="none", c="black", capsize=5
            )
        except Exception as e:
            print(
                f"Error bars aren't valid: {error[:1], error[1:]}. Skipping."
            )
        ax.text(score + error[1] + 1, i, f"{score}", va="center", color="black")

    # Remove the legend to create a shared legend later
    ax.get_legend().remove()

    # Set labels and title
    ax.set_xlabel("")
    ax.set_ylabel("")
    ax.grid(axis="x", linestyle="--")
    ax.set_xlim(right=stats["score"].max() + 20)
    ax.set_title(title)
```

```python
import matplotlib.pyplot as plt

# Load your data
gpt4_data = pd.read_json("/Users/justinzhao/Downloads/gpt4_data.json")
qwen_data = pd.read_json("/Users/justinzhao/Downloads/qwen_data.json")

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(20, 8))

# Plot each dataset on its respective Axes
plot_arena_hard_elo_stats(gpt4_data, "GPT-4-0613 as the reference model", ax=axes[0])
plot_arena_hard_elo_stats(qwen_data, "Qwen-1.5-32B as the reference model", ax=axes[1])

# Create a shared legend
handles, labels = axes[0].get_legend_handles_labels()
fig.legend(handles, labels, title="Family", loc="upper center", bbox_to_anchor=(0.5, 1.05), ncol=len(labels))

# Adjust layout and save the figure
fig.tight_layout(rect=[0, 0, 1, 0.95])  # Adjust rect to make room for the legend
# fig.tight_layout()
plt.savefig("/Users/justinzhao/Downloads/merged_plot.pdf", bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
```

```python
def plot_heatmap(
    data,
    title,
    xlabel,
    ylabel,
    vmin=None,
    vmax=None,
    center=None,
    cmap="coolwarm",
    norm=None,
    outfile=None,
    figsize=None,
    font_size=18,
):
    # Increase font size globally
    plt.rcParams.update({"font.size": font_size})

    # Convert the nested dictionary to a DataFrame
    df = pd.DataFrame(data)

    # Plotting the heatmap
    plt.figure(figsize=(20, 16) if figsize is None else figsize)
    sns.heatmap(
        df,
        annot=df,
        cmap=cmap,
        fmt=".2f",
        vmin=vmin,
        vmax=vmax,
        center=center,
        norm=norm,
    )
    # plt.title(title)
    plt.title("")
    plt.setp(plt.xticks()[1], rotation=90)
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    plt.tight_layout()

    if outfile:
        plt.savefig(outfile)
    plt.close()
```
